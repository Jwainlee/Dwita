```python
import pandas as pd
```

```python
df = pd.read_csv('/content/train.csv')
df.head()
df.info()
```

```python
df.columns[2:5]
```

```python
df_1 = df.drop(columns=df.columns[57:64])
```

```python
df_1.info()
```

```python
df_2 = df_1.copy()
df_2 = df_2.iloc[:,1:]
```

```python
df_2.head()
```

```python
import numpy as np

for i in range(df_2.shape[1]):
    m = np.mean(df_2.iloc[:, i], axis=0)
    for j in range(df_2.shape[0]):
        a = df_2.iloc[j, i]
        df_2.iloc[j, i] = m if pd.isna(a) else a
```

```python
df_2.info()
```

```python
cor = df_2.corr()
```

```python
cor = cor['emission'].abs()
```

```python
cor.sort_values(ascending = False)
```

```python
df_2.to_csv('./df.csv', index=False)
```
