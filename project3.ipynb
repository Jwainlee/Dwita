```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('/content/drive/MyDrive/data/연령.csv')
```

```python
df.head()
```

```python
df.index = df.iloc[:,0]
df = df.iloc[:,1::]
```

```python
df.head()
```

```python
for col in df.columns:
  if df[col].dtype != 'int':
    df[col] = df[col].str.replace(',', '').astype(int)
```

```python
df.info()
```

```python
df_2018 = df.iloc[0,0:12]
df_2019 = df.iloc[0,12:24]
df_2020 = df.iloc[0,24:36]
df_2021 = df.iloc[0,36:48]
df_2022 = df.iloc[0,48:60]
```

```python
df_2018.values
```

```python
df_2018.info()
```

```python
index = ['sum', 'under10', '10s', '20s', '30s', '40s', '50s', '60s', '70s', '80s', '90s', 'over100']
```

```python
df_2018.index = index
df_2019.index = index
df_2020.index = index
df_2021.index = index
df_2022.index = index
```

```python
df_2018
```

```python
labels = df_2018.index[1::]
values = df_2018.values[1::]

plt.figure(figsize=(10, 6))
plt.bar(labels, values, color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Group Distribution')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
